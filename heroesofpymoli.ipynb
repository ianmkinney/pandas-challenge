```python
# Dependencies
import os
import pandas as pd
```

```python
# Create csv_path variable
csv_path = os.path.join('HeroesOfPymoli','Resources', 'purchase_data.csv')

# Read with pandas
purchase_frame = pd.read_csv(csv_path)
purchase_frame.head()
```

```python
### Player Count
# Total Number of Unique Playes
player_count = len(purchase_frame['SN'].unique())
player_count
```

```python
### Purchasing Analysis
# Number of Unique Items
unique_items = len(purchase_frame['Item Name'].unique())
# Average Price
avg_price = round(purchase_frame['Price'].mean(), 2)
# Total Number of Purchases
purchase_total = len(purchase_frame)
# Total Revenue
total_revenue = round(purchase_frame['Price'].sum(), 2)
purchase_analysis = dict({'Unique Items':unique_items,'Average Price':f"${avg_price}",'Total Purchases':purchase_total,'Total Revenue':f"${total_revenue}"})
purchase_analysis
```

```python
### Gender Demographics
# Male Count 
gender_list = purchase_frame['Gender'].unique()
gender_list
```

```python
gender_demo = dict()
for gender in gender_list:
    tempframe = purchase_frame.loc[purchase_frame['Gender'] == gender]
    count = len(tempframe['SN'].unique())
    gender_demo[f"{gender} Count"] = count
    gender_demo[f"{gender} Percentage"] = f"{round((count/player_count*100),2)}%"
gender_demo
```

```python
def analysis(df):
    analysis = dict()
    analysis['Purchase Count'] = len(df)
    analysis['Average Purchase Price'] = f"${round(df['Price'].mean(), 2)}"
    analysis['Total Purchase Value'] = f"${round(df['Price'].sum(), 2)}"
    per_person = pd.DataFrame()
    players = df['SN'].unique()
    i = 0
    total = 0
    for player in players:
        tempframe = df.loc[df['SN'] == player]
        total = total + tempframe['Price'].sum()
        i = i + 1
    analysis['Average Purchase Total'] = f"${round(total/i,2)}"
    return analysis
```

```python
gender_breakdown = dict()
for gender in gender_list:
    tempframe = purchase_frame.loc[purchase_frame['Gender'] == gender]
    gender_breakdown[gender] =  analysis(tempframe)
gender_breakdown
```

```python
# Print out max age and min age from dataset to create correct age groups
ages = purchase_frame['Age'].unique()
print(f"Max Age: {max(ages)}, Min Age: {min(ages)}")
```

```python
### Age Demographics
# Age ranges 
age_ranges = ["0-9","10-14","15-19","20-24","25-29","30-34","35-39","40-44","45-59"]
age_demographics = dict()
for age_range in age_ranges:
    parts = age_range.split('-')
    start = int(parts[0])
    end = int(parts[1])
    tempframe = purchase_frame.loc[(purchase_frame['Age'] >= start) & (purchase_frame['Age'] <= end)] 
    age_demographics[age_range] = analysis(tempframe)
age_demographics
```

```python
### Top Spenders
# Group dataframe by SN and drop everything but Price
grouped_frame = pd.DataFrame()
grouped_frame = purchase_frame.groupby(['SN'], as_index=False).sum().sort_values(['Price'], ascending=False)
top_five = grouped_frame.head(5)
players = top_five['SN'].unique()
topfive_output = dict()
i = 1
for player in players: 
    tempdict = dict()
    tempframe = purchase_frame.loc[purchase_frame['SN'] == player]
    purchase_count = len(tempframe)
    avg_price = tempframe['Price'].mean()
    total_value = tempframe['Price'].sum()
    tempdict['SN'] = player
    tempdict['Purchase Count'] = purchase_count
    tempdict['Average Purchase Price'] = round(avg_price, 2)
    tempdict['Total Purchase Value'] = round(total_value, 2)
    topfive_output[i] = tempdict
    i = i+1
topfive_output
```
