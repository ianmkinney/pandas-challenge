```python
# Dependencies
import os
import pandas as pd
```

```python
# Create csv_path variable
csv_path = os.path.join('HeroesOfPymoli','Resources', 'purchase_data.csv')

# Read with pandas
purchase_frame = pd.read_csv(csv_path)
purchase_frame.head()
```

```python
### Purchasing Analysis
# Number of Unique Items
unique_items = len(purchase_frame['Item Name'].unique())
# Average Price
avg_price = round(purchase_frame['Price'].mean(), 2)
# Total Number of Purchases
purchase_total = len(purchase_frame)
# Total Revenue
total_revenue = round(purchase_frame['Price'].sum(), 2)
purchase_analysis = pd.DataFrame({"Unique Items": [unique_items],
                                  "Average Price": f"${avg_price}",
                                  "Total Purchases": [purchase_total],
                                  "Total Revenue": f"${total_revenue}"})
purchase_analysis
```

```python
### Top Spenders
# Group dataframe by SN and drop everything but Price
grouped_frame = pd.DataFrame()
grouped_frame = purchase_frame.groupby(['SN'], as_index=False).sum().sort_values(['Price'], ascending=False)
top_five = grouped_frame.head(5)
players = top_five['SN'].unique()
topfive_output = dict()
i = 1
for player in players: 
    tempdict = dict()
    tempframe = purchase_frame.loc[purchase_frame['SN'] == player]
    purchase_count = len(tempframe)
    avg_price = tempframe['Price'].mean()
    total_value = tempframe['Price'].sum()
    tempdict['SN'] = player
    tempdict['Purchase Count'] = purchase_count
    tempdict['Average Purchase Price'] = round(avg_price, 2)
    tempdict['Total Purchase Value'] = round(total_value, 2)
    topfive_output[i] = tempdict
    i = i+1
topfive_output
```
