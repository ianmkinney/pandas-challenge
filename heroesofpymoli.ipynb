```python
# Dependencies
import os
import pandas as pd
```

```python
# Create csv_path variable
csv_path = os.path.join('HeroesOfPymoli','Resources', 'purchase_data.csv')

# Read with pandas
purchase_frame = pd.read_csv(csv_path)
purchase_frame.head()
```

```python
### Player Count
# Total Number of Unique Playes
player_count = len(purchase_frame['SN'].unique())
player_count
```

```python
### Purchasing Analysis
# Number of Unique Items
unique_items = len(purchase_frame['Item Name'].unique())
# Average Price
avg_price = round(purchase_frame['Price'].mean(), 2)
# Total Number of Purchases
purchase_total = len(purchase_frame)
# Total Revenue
total_revenue = round(purchase_frame['Price'].sum(), 2)
purchase_analysis = pd.DataFrame({"Unique Items": [unique_items],
                                  "Average Price": f"${avg_price}",
                                  "Total Purchases": [purchase_total],
                                  "Total Revenue": f"${total_revenue}"})
purchase_analysis
```

```python
index_list = []
columns = ['Total Count', 'Percentage of Players']
data = []
gender_list = purchase_frame['Gender'].unique()
for gender in gender_list:
    tempframe = purchase_frame.loc[purchase_frame['Gender'] == gender]
    data.append([len(tempframe['SN'].unique()),f"{round(len(tempframe['SN'].unique())/player_count * 100,2)}%"])
    index_list.append(gender)
gender_demo = pd.DataFrame(columns=columns,data=data)    
gender_demo.index = index_list
gender_demo.sort_values(by='Total Count', ascending=False, inplace=True)
gender_demo
```

```python
# Create purchasing analysis function to handle repetitive calculations
def purchasing_analysis(segment):
    columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Average Purchase Total']
    segment_list = purchase_frame[segment].unique()
    data = []
    for s in segment_list:
        tempframe = purchase_frame.loc[purchase_frame[segment] == s]
        purchase_count = len(tempframe)
        person_count = len(tempframe['SN'].unique())
        avg_purchase = tempframe['Price'].mean()
        total_purchase = tempframe['Price'].sum()
        avg_purchase_pp = round(total_purchase/person_count,2)
        data.append([purchase_count,avg_purchase,total_purchase,avg_purchase_pp])
    output = pd.DataFrame(columns=columns,data=data)
    output.index = segment_list
    output.index.name = segment
    return output
```

```python
# Use function with 'Gender' segment
gender_analysis = purchasing_analysis('Gender')
gender_analysis.sort_values(by='Purchase Count', ascending=False, inplace=True)
gender_analysis
```
