```python
# Dependencies 
import pandas as pd 
import os
```

```python
# Create paths to CSVs
csvpath_schools = os.path.join('PyCitySchools', 'Resources', 'schools_complete.csv')
csvpath_students = os.path.join('PyCitySchools', 'Resources', 'students_complete.csv')
```

```python
# Create School DataFrame
school_frame = pd.read_csv(csvpath_schools)
school_frame.head()
```

```python
# Create Student DataFrame
student_frame = pd.read_csv(csvpath_students)
student_frame.head()
```

```python
### District Summary 
# Initialize DataFrame
# Create definition for easy repetition 
district_summary = pd.DataFrame()
# Total Schools = length of unique school names
total_schools = len(school_frame['school_name'].unique())
district_summary['Total Schools'] = [total_schools]
# Total students = length of unique student names
total_students = len(student_frame['student_name'])
district_summary['Total Students'] = [total_students]
district_summary['Total Students'] = district_summary['Total Students'].map("{:,}".format)
# Total buget = the sum of budget column
total_budget = school_frame['budget'].sum()
district_summary['Total School Budget'] = [total_budget]
district_summary['Total School Budget'] = district_summary['Total School Budget'].map("${:,}".format)
# Average math score = the mean of math_score
avg_math_score = student_frame['math_score'].mean()
district_summary['Average Math Score'] = round(avg_math_score,2)
# Average reading score = the mean of reading score
avg_reading_score = student_frame['reading_score'].mean()
district_summary['Average Reading Score'] = round(avg_reading_score,2)
# Percent passing math = length of passing student frame/total_students*100
pass_math_frame = student_frame.loc[student_frame['math_score'] >= 70]
passing = len(pass_math_frame['student_name'])
percent_passing_math = f"{round(passing/total_students*100,2)}%"
district_summary['Percent Passing Math'] = [percent_passing_math]
# Percent passing reading = length of passing student frame/total_students*100
pass_reading_frame = student_frame.loc[student_frame['reading_score'] >= 70]
passing = len(pass_reading_frame['student_name'])
percent_passing_reading = f"{round(passing/total_students*100,2)}%"
district_summary['Percent Passing Reading'] = [percent_passing_reading]
# Percent passing overall = length of passing student frame/total_students*100
pass_frame = student_frame.loc[student_frame['reading_score'] >= 70]
pass_frame = pass_frame.loc[pass_frame['math_score'] >= 70]
passing = len(pass_frame['student_name'])
percent_passing = f"{round(passing/total_students*100,2)}%"
district_summary['Percent Passing Overall'] = [percent_passing]
district_summary
```
