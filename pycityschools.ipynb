```python
# Dependencies 
import pandas as pd 
import os
```

```python
# Create paths to CSVs
csvpath_schools = os.path.join('PyCitySchools', 'Resources', 'schools_complete.csv')
csvpath_students = os.path.join('PyCitySchools', 'Resources', 'students_complete.csv')
```

```python
# Create School DataFrame
school_frame = pd.read_csv(csvpath_schools)
school_frame.head()
```

```python
# Create Student DataFrame
student_frame = pd.read_csv(csvpath_students)
student_frame.head()
```

```python
### District Summary 
district_summary = pd.DataFrame()
total_schools = len(school_frame['school_name'].unique())
total_students = len(student_frame['student_name'].unique())
total_budget = school_frame['budget'].sum()
avg_math_scare = student_frame['math_score'].mean()
avg_reading_score = student_frame['reading_score'].mean()
pass_math_frame = student_frame.loc[student_frame['math_score'] >= 70]
passing = len(pass_math_frame['student_name'].unique())
percent_passing_math = passing/total_students*100
fail_math_frame = student_frame.loc[student_frame['math_score'] < 70]
failing = len(fail_math_frame['student_name'].unique())
percent_failing_math = failing/total_students*100

```
