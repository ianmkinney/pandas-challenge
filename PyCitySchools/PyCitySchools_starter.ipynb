```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_frame = pd.read_csv(school_data_to_load)
student_frame = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_frame, school_frame, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
### School Summary
# Create definition that creates a summary for the school passed
school_names = school_data_complete['school_name'].unique()
# Create list of column names
columns = ['School Name','School Type', 'Total Students', 'Total School Budget', 'Per Student Budget', 'Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']
# Create empty list to store dataframes for concatenation
dataframes = []
def school_summary(school): 
    # Create temporary dataframe with current school's data
    tempframe = school_data_complete.loc[school_data_complete['school_name'] == school]
    # Define and fill Variables
    school_type = tempframe['type'].unique()[0]
    total_students = len(tempframe['student_name'])
    total_school_budget = round(tempframe['budget'].sum(),2)
    per_student_budget = round(total_school_budget/total_students,2)
    average_math_score = round(tempframe['math_score'].mean(), 2)
    average_reading_score = round(tempframe['reading_score'].mean(), 2)

    # Percent passing math = length of passing student frame/total_students*100
    pass_math_frame = tempframe.loc[student_frame['math_score'] >= 70]
    passing = len(pass_math_frame['student_name'])
    percent_passing_math = f"{round(passing/total_students*100,2)}%"

    # Percent passing reading = length of passing student frame/total_students*100
    pass_reading_frame = tempframe.loc[student_frame['reading_score'] >= 70]
    passing = len(pass_reading_frame['student_name'])
    percent_passing_reading = f"{round(passing/total_students*100,2)}%"

    # Percent passing overall = length of passing student frame/total_students*100
    pass_frame = tempframe.loc[student_frame['reading_score'] >= 70]
    pass_frame = pass_frame.loc[pass_frame['math_score'] >= 70]
    passing = len(pass_frame['student_name'])
    percent_passing = f"{round(passing/total_students*100,2)}%"

    Outdata = [[school, school_type,total_students,total_school_budget,per_student_budget,average_math_score,average_reading_score,          percent_passing_math,percent_passing_reading,percent_passing]]

    outframe = pd.DataFrame(columns=columns, data=Outdata)
    dataframes.append(outframe)
    return

# Loop through school names and use definition to fill dataframe list 
for name in school_names: 
    school_summary(name)

school_summary_mast = pd.DataFrame(columns=columns)

school_summary_mast = pd.concat(dataframes, axis=0, ignore_index=True)

school_summary_mast.head()
```
